```python
import tensorflow as tf
import numpy as np
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]], dtype=np.float32)

y = np.array([[0],
              [1],
              [1],
              [0]], dtype=np.float32)
class FFNN(tf.keras.Model):
    def __init__(self):
        super(FFNN, self).__init__()
        self.d1 = tf.keras.layers.Dense(8, activation='relu')
        self.d2 = tf.keras.layers.Dense(1, activation='sigmoid')
    def call(self, x):
        x = self.d1(x)
        return self.d2(x)
model = FFNN()
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
model.fit(X, y, epochs=100, verbose=1)
print(model(X))
```
